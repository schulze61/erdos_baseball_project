```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./heat_map_data/heat_map_data.csv')
```

```python
fig, ax = plt.subplots(figsize = (10,10))

#Number of levels for the contour plot. This particular number gives a contour interval that is approximately 
#3 times the 1-sigma variability for each data point used in the grid
levels = 28

#plot grid points
plt.plot(data['eps1'], data['eps2'], 'ko')
plt.tricontourf(data['eps1'], data['eps2'], data['bscore'], levels = levels, cmap = "RdBu_r")
plt.tick_params(which = 'both', labelsize = 18, length = 15, direction = 'in', width = 3, top = True, right = True)
cbar = plt.colorbar(label = 'Brier Score')
cbar.set_label(label='Brier Score', size = 18)
cbar.ax.tick_params(labelsize=18)


ax.patch.set_edgecolor('black')  
ax.patch.set_linewidth('3')

plt.xlabel(r'$\epsilon_{1}$', fontsize = 24)
plt.ylabel(r'$\epsilon_{2}$', fontsize = 24)

#default eps1 and eps2 for 2010 season
plt.plot([1.0], [1.4237], 'ks', markersize = 15)


```
